1. (0028, 0030) Pixel Spacing                       DS: ['0.732421875', '0.732421875']

2. (0020, 0032) Image Position (Patient)            DS: ['-180.72742462158', '-227.5', '64.1605377197265']

3. (0020, 0037) Image Orientation (Patient)         DS: ['1', '0', '0', '0', '1', '0']

4. (0018, 0050) Slice Thickness                     DS: '7'

5. (0020, 1041) Slice Location                      DS: '112.000003814697'


1. **Pixel Spacing (0028,0030)** specifies the physical distance in the patient between the center of each two-dimensional pixel, specified by two numeric values. It values account for assumed or known geometric magnification effects and correspond to some unspecified depth within the patient and thus be used for measurements of objects located close to the central ray and at the same depth.

2. **Image Position (Patient) (0020, 0032)** specifies the x, y, and z coordinates of the upper left hand corner of the image and it is the center of the first voxel transmitted.

3. **Image Orientation (Patient) (0020, 0037)** specifies the direction cosines of the first row and the first column with respect to the patient. These Attributes is provided as a pair. Row value for the x, y, and z axes is followed by the Column value for the x, y, and z axes respectively.The direction of the axes is defined fully by the patient's orientation.

4. **Slice Thickness (0018, 0050)** specifies nominal slice thickness in mm.

5. **Slice Location (0020, 1041)** is defined as the relative position of the image plane expressed in mm. It is relative to an unspecified implementation specific reference point.

Quantitative computed tomography(QCT) is a 3D method obtained on a CT scanner that provides true bone density by detecting low bone mass. By focusing current CT and resulted QCT image we can convert the image information into bone density data. Apart from this, the cone beam computerized tomography (CBCT) images could provide reliable information regarding bone defects in the any region but no systematic information regarding the standardization of CBCT and properties is available and also there is a lack of information on how the images are generated, exported, and analyzed.For the evaluation of bone defects, measurements of bone volume, height, width, and density are required.

DICOM addresses five areas: (1) network image management, (2) network image interpretation management, (3) network print management, (4) imaging procedure management, (5) off-line storage media management. It specifies data interchange protocol, digital image format, and file structure for biomedical images and image-related information.

The attributes obtained using im140.dcm DICOM File are very restricted. It is really hard to preprocess data to generate a histogram and niether do we have other important feature like Image Orientation. Hence, the actual bone intenstiy information is a monotonous job. The only attributes with us are **Pixel Spacing, Samples/Pixel, Revolution Time, Slice Thickness, Series Time and Acquisition Time**. Using these information, we can derive a hypothetical model of the obtained DICOM because we have the time duration of the samples resulted. We have C/C distance between two slices and the series and derivation period of each slice.
