Syntax of equalize_hist:-
--------------------------

* **skimage.exposure.equalize_hist(image, nbins, mask)**

Parameters:-
-------------
* **image** : array
<br>Image array.
* **nbins** : int, optional
<br>Number of bins for image histogram.
* **mask**: ndarray of bools or 0s and 1s, optional
<br>Same shape of array as image. If mask= true used for equalization than it is applied to whole image

Returns:-
----------
* **out**: float array
<br>Image array after histogram equalization.

Explanation of the function equalize_hist :-
----------------------------------------------
**Histogram equalization is a technique for adjusting image intensities to enhance contrast and exposure.** It **spread the pixel values uniformly in the histogram due to which specific portion under new histogram increases linearly from left to right and has the practical effect of making a sharper, contrast-enhanced image.** It actually redistributes the value densities along the histogram so that each value range is more evenly represented and the **valleys are compressed while peaks in histogram are widened.**

The **benefits using equalize histogram is it does not require parameters and sometimes provide images which is like unnatural.** For instance, all pixels values are restricted to high values in brighter image while good image will have pixels from all area of image. Thus, by stretching ends of histogram it will improve the contrast of the image

Syntax of equalize_adapthist :-
----------------------------------
* **skimage.exposure.equalize_adapthist(image, kernel_size, clip_limit, nbins)**

Parameters:-
-------------
* **image** : (M, N[, C]) ndarray
<br>Input image
* **kernel_size**: integer or list-like, optional
<br>It defines the shape of contextual regions used in the algorithm. 
* **nbins** : int, optional
<br>Number of gray bins for histogram.

Returns :-
-----------
* **out** : (M, N[, C]) ndarray
<br>Equalized image.

Explanation of the function equalize_adapthist :-
----------------------------------------------------
This is an extension to traditional Histogram Equalization technique. **It enhances the contrast of images by transforming the values in the intensity image. It computes several histograms each corresponding to a distinct section of the image and uses them to redistribute the lightness values of the image.** It is an algorithm for local contrast enhancement, that uses histograms computed over different regions of the image. Even in regions that are darker or lighter local details can be enhanced . **It is therefore suitable for improving the local contrast and enhancing the definitions of edges in each region of an image.**

**It has a tendency to overamplify noise in relatively homogeneous regions of an image. Advance adaptive histogram known by contrast limited adaptive histogram equalization (CLAHE) prevents it by limiting amplification.** It splits the image into small rectangular areas called tiles and enhances the contrast of these areas by adjusting their local histograms.

**Adaptive histogram equalization can be used when presence of noise is less because it is used to operate on small data regions (tiles) rather than the entire image which makes to maintain roughly the same histogram shape without equalizing noise. **Each tile's contrast is enhanced so that the histogram of each output region approximately matches the specified histogram which **results in uniform distribution.** The main **advantage** using adaptive histogram equalization is that the **contrast enhancement can be limited in order to avoid amplifying the noise which might be present in the image.**

While **standard histogram equalizaton which is also known as Global histogram because it enhances the whole image and makes it intensities to be extra bright.** Therefore, global histogram is **used when there is less noise in the image. It seems to be less sensitive to noise but does not improve the contrast of subtle changes by much. On the other hand, adaptive histogram is more sensitive to noise but is capable of enhancing subtle changes in the region for image processing purposes.**

Hereby, it can be concluded that **Adaptive histogram equalization is more useful in the notebook than standard histogram equalization because it enhance each areas of local neighbourhood pixels which results into better visibility of image and such type of outcome is not possible in normal equalization beacuse it will contrast the whole image with extra brightness.**
