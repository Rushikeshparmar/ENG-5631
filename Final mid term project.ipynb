Introduction on ECG & ECG signals :-
--------------------------------------

**Electrocardiogram(ECG)** is an vital diagnosis of heart diseases. It is a signal which is time varying and  reflects ionic current flow thus results cardiac fibres to contract and relax. The electrical activity of heart is due to rapid ionic movements from membranes which causes Depolarization and repolarization of myocardial cell. It is also used to obtain heartbeat by calculation of time between QRS complex. It  amplifies and detects small electrical changes on surface of skin due to depolarization of heart muscle during each heart beat. A one cycle of ECG represents atrial or ventricular depolarisation/repolarisation which occurs with every heartbeat. Each cardiac cycles consists of P-QRS-T signals. Electrical activity in the heart facilitate detection from the outer body.


**There are four stages in ECG signal processing:-**

**1.Data Acquisition-** To encode signal and store in suitable readable computer form.

**2.Signal Conditioning-**To eliminate extraneous components like noise using different filters.

**3.Feature extraction-**To identify and measure number of parameters that best characterize desired information.

**4.Hypothesis Testing-**Statistical techniques used to answer question


ECG signal can be seek by recording potential difference between two electrodes figured on surface of skin by using 12 lead recording system which are attached at 12 different points. These signal is passed to amplifier for coupling output and cancellation extraneous component like noise thereafter it acts as input to lowpass & highpass filter to block frequency components. The signal is then rectified by main amplifier. The function of heart is regulated by amplitude and interval of P-QRS-T segment. QRS complex is obtained in feature extraction which includes R-peak detection, QT-peak detection, derivation of heart, respiration rate  and RR interval it is time between successive R-peaks as well as inverse of this time gives heart rate. Where,

P-wave represents Atria contraction

QRS complex represents ventricular contraction & depolarization

T-wave represents ventricular repolarization

S-T segment where T&U wave together represents total duration of ventricular recovery

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

fs = 750.0 #Sampling frequency
lowcutoff=100.0 #Low cuttoff frequency
nyq=0.5*fs #Nyquist limit
data=lowcutoff/nyq
print(data)

#Define filter signal
b, a = signal.butter(4, data , btype='lowpass', analog=False)
w, h = signal.freqz(b, a)

#Plotting Magnitude response
plt.plot(w, 20 * np.log10(abs(h)))
plt.title('Butterworth Filter ')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.grid(which='both', axis='both')
plt.xlabel('log')
# plt.axvline(100, color='green') # cutoff frequency
plt.show()
```

```python
#Filtering the signal with Butterworth filter

from scipy import signal
import scipy as Sci
import numpy as np
import matplotlib.pyplot as plt


xsign = AllSignal[:43081]-np.mean(AllSignal[:43081])
x=np.arange(0,43081/750.,1/750.) 
b, a = signal.butter(4, data , btype='lowpass', analog=False)
#filtering signal
outputsignal=signal.lfilter(b,a,xsign)

#plotting input signal
plt.subplots_adjust( hspace = .5 )
plt.subplot(2, 1, 1)
plt.title('Input signal')
plt.plot(x,AllSignal[:43081])
plt.xlim(0,0.5)
plt.xlabel('Time')
plt.ylabel('Amplitude')
##plotting major & minor grid
plt.grid(b=True, which='major', color='k', linestyle='-')
plt.grid(b=True, which='minor', color='r', linestyle='-', alpha=0.2)
plt.minorticks_on()
plt.show()

#plotting output filtered signal
plt.subplot(2, 1, 2)
plt.title('Output Filtered signal')
plt.plot(x,outputsignal)
plt.xlim(0,0.5)
plt.xlabel('Time')
plt.ylabel('Amplitude')
#plotting major & minor grid
plt.grid(b=True, which='major', color='k', linestyle='-')
plt.grid(b=True, which='minor', color='r', linestyle='-', alpha=0.2)
plt.minorticks_on()
plt.show()
```
